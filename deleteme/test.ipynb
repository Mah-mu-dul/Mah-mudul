```python
def apply_type_3_effect(query_result, string):
    X = query_result
    modified_string = ""
    for char in string:
        if char.isalpha():
            modified_char = chr(((ord(char) - ord('a') + X) % 26) + ord('a'))
            modified_string += modified_char
        else:
            modified_string += char
    return modified_string

def solve_test_case(queries):
    B = []
    X = 0
    result = []

    for query in queries:
        query_type, *args = query.split()
        
        if query_type == "1":
            S, f = args
            f = int(f)
            modified_S = apply_type_3_effect(X, S)
            B.extend([modified_S] * f)
        elif query_type == "2":
            S, f = args
            f = int(f)
            modified_S = apply_type_3_effect(X, S)
            B = [string for string in B if string != modified_S]
            # Alternatively: B = [string for string in B if string != S]
        elif query_type == "3":
            S1, S2 = args
            sorted_B = sorted(B)
            idx_S1 = sorted_B.index(apply_type_3_effect(X, S1))
            idx_S2 = sorted_B.index(apply_type_3_effect(X, S2))
            result.append(abs(idx_S2 - idx_S1) + 1)
            X = result[-1]

    return result

def main():
    T = int(input().strip())
    for _ in range(T):
        Q = int(input().strip())
        queries = [input().strip() for _ in range(Q)]
        results = solve_test_case(queries)
        for res in results:
            print(res)

if __name__ == "__main__":
    main()
```
